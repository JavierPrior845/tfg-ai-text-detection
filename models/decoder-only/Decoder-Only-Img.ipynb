```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch
import numpy as np

model_id = "unsloth/meta-llama-3.1-8b-bnb-4bit" # Versión ligera para Colab

tokenizer = AutoTokenizer.from_pretrained(model_id)
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    device_map="auto",
    dtype=torch.bfloat16
)
```

# 3. El Motor de Cálculo: Perplejidad y Logprobs
Este script procesa el texto y extrae la probabilidad de cada token para calcular la métrica de "sorpresa"

```python
def analyze_text_statistics_advanced(text):
    inputs = tokenizer(text, return_tensors="pt").to("cuda")
    input_ids = inputs["input_ids"]

    with torch.no_grad():
        outputs = model(input_ids, labels=input_ids)
        logits = outputs.logits

    shift_logits = logits[..., :-1, :].contiguous()
    shift_labels = input_ids[..., 1:].contiguous()

    # Probabilidades y Ranking
    probs = torch.nn.functional.softmax(shift_logits, dim=-1)
    ranks = torch.argsort(probs, dim=-1, descending=True)

    # Encontrar la posición del token real en el ranking
    # Esto es computacionalmente costoso, se puede optimizar
    actual_ranks = []
    for i in range(shift_labels.size(1)):
        rank = (ranks[0, i] == shift_labels[0, i]).nonzero(as_tuple=True)[0].item()
        actual_ranks.append(rank + 1) # Rank 1 es el más probable
        #actual_ranks = (ranks[0] == shift_labels[0].unsqueeze(-1)).nonzero()[:, 1] + 1 #vectorizar el calculo del ranking para ganar velocidad

    return {
        "perplexity": torch.exp(outputs.loss).item(),
        "avg_rank": np.mean(actual_ranks),
        "log_rank": np.mean(np.log(actual_ranks)),
        "entropy": -(probs * torch.log(probs + 1e-9)).sum(dim=-1).mean().item()
    }
```

# 4. Clasificación por Umbral (Thresholding)
Para el umbral vamos a calcular la curva ROC

```python
def classify_news(stats, threshold_ppl=15.0, threshold_surprise=0.15):
    # Lógica Senior: Si la perplejidad es alta o hay mucha sorpresa, es humano
    is_human_ppl = stats["perplexity"] > threshold_ppl
    is_human_surprise = stats["surprise_ratio"] > threshold_surprise

    # Podemos usar una combinación (Votación)
    prediction = "Humano" if is_human_ppl or is_human_surprise else "IA"
    return prediction
```

**Análisis de Ranks (CDF Plot)**
Este gráfico es crucial porque visualiza la "**Predecibilidad**". Si la curva roja (IA) sube mucho más rápido que la azul (Humano), significa que Gemini elige palabras que están casi siempre en el Top-1 o Top-2 del vocabulario, mientras que el humano es más errático.

```python
def plot_rank_cdf(df_results):
    plt.figure(figsize=(10, 6))

    # Separamos por clase
    synthetic = df_results[df_results['is_real'] == 0]['avg_rank']
    human = df_results[df_results['is_real'] == 1]['avg_rank']

    # Graficamos la CDF (Empirical Cumulative Distribution Function)
    sns.ecdfplot(data=synthetic, label='SYNTHETIC (Gemini)', color='crimson', linewidth=2.5)
    sns.ecdfplot(data=human, label='REAL (Human)', color='royalblue', linewidth=2.5)

    plt.title('Cumulative Distribution Function (CDF) of Ranks', fontsize=14, fontweight='bold')
    plt.xlabel('Average Rank per Headline (Lower = More Predictable)')
    plt.ylabel('Cumulative Probability')
    plt.legend(title="Class")
    plt.grid(True, which="both", ls="--", alpha=0.5)

    # Zoom en los primeros 10 ranks, donde ocurre la magia
    plt.xlim(1, 500)

    plt.tight_layout()
    plt.savefig("rank_cdf_analysis.png")
    plt.show()

# Ejecutar:
plot_rank_cdf(df_results)
```

```python
from sklearn.metrics import roc_curve, auc, precision_recall_curve
import matplotlib.pyplot as plt
import numpy as np

def analyze_roc_and_find_threshold(df, metric_col='perplexity', target_col='is_real'):
    """
    Calcula la curva ROC y encuentra el umbral óptimo usando el Índice de Youden.
    Nota: target_col debe ser 1 para 'Real' y 0 para 'IA'.
    """
    y_true = df[target_col]
    y_scores = df[metric_col] # Usamos la métrica como puntuación de "realidad"

    # 1. Calcular FPR, TPR y Umbrales
    fpr, tpr, thresholds = roc_curve(y_true, y_scores)
    roc_auc = auc(fpr, tpr)

    # 2. Encontrar el umbral óptimo (Índice de Youden: J = TPR - FPR)
    # Buscamos el punto que maximice la distancia a la línea diagonal
    optimal_idx = np.argmax(tpr - fpr)
    optimal_threshold = thresholds[optimal_idx]

    # 3. Visualización
    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.scatter(fpr[optimal_idx], tpr[optimal_idx], color='red', label=f'Umbral Óptimo: {optimal_threshold:.2f}')

    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate (IA clasificada como Humano)')
    plt.ylabel('True Positive Rate (Humano clasificado como Humano)')
    plt.title('Receiver Operating Characteristic (ROC)')
    plt.legend(loc="lower right")
    plt.grid(alpha=0.3)
    plt.savefig('roc_curve_analysis.png')

    print(f"AUC-ROC: {roc_auc:.4f}")
    print(f"Mejor Umbral detectado para {metric_col}: {optimal_threshold:.4f}")

    return optimal_threshold

# Uso:
best_threshold = analyze_roc_and_find_threshold(df_results)
```

```python
# Aplicamos el umbral óptimo encontrado
df_results['final_pred'] = (df_results['perplexity'] > best_threshold).astype(int)

# Generamos el reporte final de métricas
from sklearn.metrics import classification_report
print(classification_report(df_results['is_real'], df_results['final_pred'], target_names=['IA (Synthetic)', 'Humano (Real)']))
```
